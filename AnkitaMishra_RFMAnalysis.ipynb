```python
a=1
df1=df1.drop(['ORDERDATE'], axis=1)
plt.figure(figsize=(15,100))
for i in df1.columns :
    if df1[i].dtypes != 'object':
        plt.subplot(18,2,a)
        sns.distplot(df1[i])
        plt.title("Histogram for: " +i,fontsize = 16)
        plt.subplot(18,2,a+1)
        sns.boxplot(df1[i])
        plt.title("Boxplot for: " +i,fontsize = 16)
        a+=2
plt.tight_layout()
```

```python
def univaranalysis(column):
    print("Description of " + column)
    print("----------------------------------------------------------------------------")
    print('null value ',df1[column].isnull().sum())
    print(df1[column].describe(),end='     ')
    print('')
    print('')
```
